```python
import torch
from scapy.utils import CustomScheduler

# Initialze lr scheduler with any optimizer and learnable parameter.
p = torch.nn.Parameter(torch.empty(4, 4))
optimizer = torch.optim.Adam([p], lr=1e-3)

scheduler_name = 'time'

ax = None
for scheduler in ['cosine', 'time', 'step', 'exponential', 'linear']:
    scheduler = CustomScheduler(optimizer = optimizer, 
                                scheduler = scheduler,
                                warmup_epochs = 0)
    
    ax = scheduler.plot_lr_factor(epochs = 40, legend = True,
                                  return_ax = True, ax = ax)
```
